```python
import pandas as pd

house_rent = pd.read_excel("houses_rent.xlsx")

house_rent.head()
```

```python
house_rent.tail()
```

```python
house_rent = house_rent.rename(columns={
    'parking spaces': 'parking_spaces',
    'rent amount': 'rent_amount',
    'property tax': 'property_tax',
    'fire insurance': 'fire_insurance'})

#altera o nome das colunas no dataset
```

```python
house_rent.head()
```

```python
house_rent.dtypes #exibe os tipos de dados daquela tabela
```

```python
house_rent = house_rent.astype({ #altera os tipos dos dados, nesse caso de nmero inteiro para decimal
    'rent_amount': 'float',
    'property_tax': 'float',
    'fire_insurance': 'float',
    'total': 'float'
    })

pd.options.display.float_format = '{:,.2f}'.format #formata os valores float para mostrar s 2 casas decimais
```

```python
house_rent.dtypes
```

```python
house_rent.info() #mostra uma informao completa sobre o seu dataset
```

```python
house_rent.describe() #informaes mais detalhadas sobre a tabela

#std - desvio padro, responsvel por calcular a disperso dos dados (se ele for maior que a mdia e porque existem dados muito dispersos)

#As porcentagens levam em considerao o valor mnimo e mximo
#25% seria de area seria 58, ou seja, se voc quiser um apartamento com uma area entre 10(mnimo) e 58 voc ter 25% das opes que no caso seria 1520 imveis
```

```python
house_rent = house_rent[(house_rent['total'] <= 4000) & (house_rent['animal'] == "accept")] #O & estabelece uma condio igual o "and" e o | de "or"
house_rent.head()
```

```python
#Existe a funo .query() que facilita ainda mais esse procedimento
house_rent = house_rent.query("rooms == 2 or rooms == 3")

house_rent.head()
```

```python
house_rent.describe() #dataset filtrado
```

```python
# Agrupamento de Dados

#O programa est agrupando a mdia dos valores por cidade (a parte final  so para formatar a exibio do resultado)
agg_city_price = house_rent.groupby(['city'])['total'].agg("mean").reset_index()
agg_city_price
```

```python
#Agrupamento da mdia dos preos por banheiro e quarto

agg_bath_room_price = house_rent.groupby(['bathroom','rooms'])['total'].agg("mean").reset_index()
agg_bath_room_price
```

```python
import seaborn as sns

ax = sns.barplot(data=agg_bath_room_price, x="rooms", y="total", color="purple"); # ; para no mostrar o cdigo no grfico
ax.set(xlabel = "Qtd Quartos", ylabel = "Mdia de Preos");
```

```python
ax = sns.barplot(data=agg_bath_room_price, x="bathroom", y="total", color="yellow");
ax.set(xlabel="Qtd Banheiros", ylabel="Mdia dos Preos");
```
